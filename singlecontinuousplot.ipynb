```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.dates as mdates
d = pd.read_csv("weekdec.csv")
x = pd.to_datetime(d.DateTime, format='%d-%m-%Y %H:%M')
x_numeric = mdates.date2num(x) 
y = d.Inverter01MeasuredDCVoltage1
z = d.Inverter01MeasuredDCCurrent1

fig = plt.figure(figsize=(10, 10)) 
ax = fig.add_subplot(111, projection='3d')
ax.grid()
ax.plot3D(x_numeric, y, z)
ax.set_title('3D Plot')
ax.set_ylabel('Inverter-01 - Measured DC Voltage 1, avg (V)', labelpad=30)
ax.set_zlabel('Inverter-01 - Measured DC Current 1, avg (A)', labelpad=30)
ax.set_xlabel('DateTime', labelpad=30)
days = mdates.DayLocator(interval=1) 
day_fmt = mdates.DateFormatter('%m-%d') 
ax.xaxis.set_major_locator(days)
ax.xaxis.set_major_formatter(day_fmt)

plt.show()
```
