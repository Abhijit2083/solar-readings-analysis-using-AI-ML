```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
d=pd.read_csv("sone.csv")
x=d.TranslatedVoltage
y=d.TranslatedCurrent
def pmax():
    max_product = 0 
    
    for i in range(len(x)):
        product = x[i] * y[i]
        
        if product > max_product:
            max_product = product
            r=x[i]
            s=y[i]
            
    return max_product,r,s
p,tv,tc=pmax()
plt.plot(d.TranslatedVoltage,d.TranslatedCurrent,color='blue',label='Original curve')
plt.xlabel("Translated Voltage (Volts)")
plt.ylabel("Translated Current (Amps)")
plt.title("Voltage vs Current")
data = {
    'TV': [23.21, tv, 779.92],
    'TC': [9.08, tc, 0.13],
    'P' : [210.6538435,p,101.4348149]
}
df = pd.DataFrame(data)
x=df.TV
y=df.TC
z=df.P
x = np.array(x).reshape(-1, 1)
y = np.array(y)
z = np.array(z)
degree = 4 
poly_features = PolynomialFeatures(degree=degree)
x_poly = poly_features.fit_transform(x)
model = LinearRegression()
model.fit(x_poly,y)
voltage_range = np.linspace(min(x), max(x), 100).reshape(-1, 1)
voltage_range_poly = poly_features.transform(voltage_range)
predicted_current = model.predict(voltage_range_poly)
plt.scatter(x, y, color='blue', label='Sample points')
plt.plot(voltage_range, predicted_current, color='red', label='Predicted curve')
plt.xlabel('Voltage')
plt.ylabel('Current')
plt.title('Current vs Voltage')
plt.legend()
plt.show()
```

```python
def v():
    arr=[]
    for i in voltage_range:
        arr.append(i)
        i=i+1
    return arr
def i():
    arr1=[]
    for i in predicted_current:
        arr1.append(i)
        i=i+1
    return arr1
vf=v()
ir=i()
print(vf)
print(ir)
```
