```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
from mplcursors import cursor

d=pd.read_csv("sone.csv")
x=d.TranslatedVoltage
y=d.TranslatedCurrent
plt.plot(d.TranslatedVoltage,d.TranslatedCurrent,color='red')
plt.xlabel("Translated Voltage (Volts)")
plt.ylabel("Translated Current (Amps)")
plt.title("Voltage vs Current")
def pmax():
    max_product = 0 
    
    for i in range(len(x)):
        product = x[i] * y[i]
        
        if product > max_product:
            max_product = product
            r=x[i]
            s=y[i]
            
    return max_product,r,s
p,tv,tc=pmax()
print("Pmax = ",p)
print("Translated Voltage (Volts) = ",tv)
print("Translated Current (Amps) = ",tc)
plt.plot(tv,tc,marker = 'o')
cursor(hover=True)
plt.show()
```
